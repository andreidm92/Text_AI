1. Создайте мешок слов с помощью sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed' и 'tweet_lemmatized' отдельно.
Игнорируем слова, частота которых в документе строго превышает порог 0.9 с помощью max_df.
Ограничим количество слов, попадающий в мешок, с помощью max_features = 1000.
Исключим стоп-слова с помощью stop_words='english'. 
Отобразим Bag-of-Words модель как DataFrame. columns необходимо извлечь с помощью CountVectorizer.get_feature_names().

```python
from sklearn.feature_extraction.text import CountVectorizer

```

```python
unpickled_list['tweet_stemmed'] = unpickled_list['tweet_stemmed'].apply(lambda x: " ".join(x) )
unpickled_list['tweet_stemmed']
```

```python
unpickled_list['tweet_lemmatized'] = unpickled_list['tweet_lemmatized'].apply(lambda x: " ".join(x) )
unpickled_list['tweet_lemmatized']
```

```python
unpickled_list.head()
```

Мешок слов для tweet_stemmed CountVectorizer

```python
count_vectorizer = CountVectorizer(ngram_range=(1, 1), analyzer='word', lowercase=True, binary=True, stop_words='english', max_df= 0.9, max_features = 1000)
bag_of_words = count_vectorizer.fit_transform(unpickled_list['tweet_stemmed'])
feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
```

Мешок слов для tweet_lemmatized CountVectorizer

```python
count_vectorizer = CountVectorizer(ngram_range=(1, 1), analyzer='word', lowercase=True, binary=True, stop_words='english', max_df= 0.9, max_features = 1000)
bag_of_words = count_vectorizer.fit_transform(unpickled_list['tweet_lemmatized'])
feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
```
